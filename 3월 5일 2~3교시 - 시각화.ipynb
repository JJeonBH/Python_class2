```python
# 라인 그래프 그리기
# 그릴 데이터 점수 : 40, 50, 70, 90
sungjuk = [40, 10, 70, 45]
x = ['kildong', 'younghee', 'chulsoo', 'miyoung']
plt.plot(x, sungjuk, 'ro') # 마커 추가
plt.show()
```

```python
sungjuk = [40, 50, 70, 90]
x = [1, 2, 3, 4]

plt.plot(x, sungjuk, 'bo') # 마커 추가
plt.plot(x, sungjuk) # 라인그래프

# 좌표의 최대/최소값 설정[xmin, xmax, ymin, ymax]
plt.axis([0, 5, 0, 100])
plt.show()
```

```python
# 막대 그래프 그리기
y = [40, 50, 70, 90]
x = ['kildong', 'younghee', 'chulsoo', 'miyoung']
plt.bar(x, y)
plt.show()
```

```python
# 막대 그래프 그리기
y = [160, 170, 155, 190, 180]
x = ['kildong', 'younghee', 'chulsoo', 'miyoung', 'minsoo']
plt.bar(x, y)
plt.plot(x, y)
plt.axis([-1, 5, 100, 200])
```

```python
# 산점도 (데이터 분포를 그린다.)
np.random.seed(0)
n = 50
x = np.random.rand(n)
y = np.random.rand(n)
area = (100 * np.random.rand(n)) * 2
colors = np.random.rand(n)
plt.scatter(x, y, s = area, c = colors)
```

```python
# 파이 차트(백분율로 보여주기)
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.show()
```

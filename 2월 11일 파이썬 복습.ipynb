```python
# 성적 처리 프로그램


# 성적 입력(점수 : 0~100)
# 점수 범위 체크
def insert_check():
    while True:
        try:
            kor = int(input('국어 : '))
        except:
            print('다시 입력하세요.')
            continue
        if kor < 0 or kor > 100:
            print('다시 입력하세요')
            continue
        break
        
    while True:
        try:
            eng = int(input('영어 : '))
        except:
            print('다시 입력하세요.')
            continue
        if eng < 0 or eng > 100:
            print('다시 입력하세요')
            continue
        break
        
    while True:
        try:
            math = int(input('수학 : '))
        except:
            print('다시 입력하세요.')
            continue
        if math < 0 or math > 100:
            print('다시 입력하세요')
            continue
        break
            
    kor_eng_math = [kor, eng, math]
            
    return kor_eng_math


# 성적 수정
# 점수 범위 체크
def update_insert_check(student, name, subject):
    if subject == '국어':
        while True:
            try:
                kor = int(input('국어 : '))
            except:
                print('다시 입력하세요.')
                continue
            if kor < 0 or kor > 100:
                print('다시 입력하세요')
                continue
            student[name][0] = kor
            break
    if subject == '영어':
        while True:
            try:
                eng = int(input('영어 : '))
            except:
                print('다시 입력하세요.')
                continue
            if eng < 0 or eng > 100:
                print('다시 입력하세요')
                continue
            student[name][1] = eng
            break
    if subject == '수학':
        while True:
            try:
                math = int(input('수학 : '))
            except:
                print('다시 입력하세요.')
                continue
            if math < 0 or math > 100:
                print('다시 입력하세요')
                continue
            student[name][2] = math
            break
                
    kor_eng_math = [ student[name][0], student[name][1], student[name][2] ]
    
    return kor_eng_math


# 성적 받아서 총점, 평균 구하기
def add_avg(kor_eng_math):
    score = 0
    avg = 0
    for i in kor_eng_math:
        score += i
        
    avg = round(score / 3, 2) # 반올림 함수
    
    score_avg = [score, avg]
    
    return score_avg


# -- 성적 입력 함수 --
def insert_sungjuk(student):
    name = input('이름 입력 : ')
    
    if(name not in student) == True:
        
        # 성적 입력(점수 : 0~100)
        # 점수 범위 체크
        kor_eng_math = insert_check()
        print('성적입력이 되었습니다.')
        
        # 성적처리를 함수로 만들기도 고려해보자
        score_avg = add_avg(kor_eng_math)
    else:
        print('학생이 이미 있습니다.')
        return insert_sungjuk(student)
    
    # student 저장소에 담아서 전달
    student[name] = kor_eng_math + score_avg
    return student


# -- 성적 출력 함수 --
def show(student):
    print('===================================')
    print('과목점수:국어 영어 수학 총점 평균')
    
    # 딕셔너리(키와 값)의 값을 하나씩 출력
    for key, value in student.items():
        print(key, ':', value)
        
    print('===================================')
    

# -- 성적 검색 함수 --
def find_by_name(student):
    # 검색할 이름 사용자로부터 받음
    name = input('검색할 이름 : ')

    # student 저장소에서 검색
    if(name in student) == True:
        print('-----------------------------------')
        print('점수 : 국어, 영어, 수학, 총점, 평균')
        print(name, ':', student.get(name),' ')
        print('-----------------------------------')
    else:
        print('해당 이름이 없습니다.')
        print('다시 검색하려면 1번 아니면 메인 화면으로 되돌아 갑니다.')
        num = input('번호 입력[1] : ')
        if num == '1':
            find_by_name(student)
        else:
            return
    return


# -- 성적 수정 함수 --
def update(student):
    name = input('수정할 학생이름 : ')

    # 입력한 이름이 존재하는지 확인
    if(name in student) == True:
        while True:
            subject = input('수정할 과목 입력 : ')
            kor_eng_math = update_insert_check(student, name, subject)
            print('수정할 과목이 더 있나요? 있으면 1번 아니면 메인 화면으로 되돌아 갑니다')
            num = input('번호 입력[1] : ')
            if num == '1':
                continue
            else:
                break
        
        score_avg = add_avg(kor_eng_math)
        
        student[name] = kor_eng_math + score_avg
    else:
        print('해당 이름의 데이터가 없습니다.')
        print('다시 수정하려면 1번 아니면 메인 화면으로 되돌아 갑니다.')
        num = input('번호 입력[1] : ')
        if num == '1':
            update(student)
        else:
            return
    return student


# -- 성적 삭제 함수 --
def delete(student):
    while True:
        name = input('삭제할  학생의 이름 : ')

        # pop(키) : 삭제
        if(name in student) == True:
            student.pop(name)
            print(name + '이(가) 삭제 되었습니다.')
            break
        else:
            print('해당 이름의 데이터가 없습니다.')
            print('다시 삭제하려면 1번 아니면 메인 화면으로 되돌아 갑니다.')
            num = input('번호입력[1] : ')
            if num == '1':
                continue
            else:
                break
    
    return student
    
    
    
    
# student 딕셔너리 생성
student = dict() # 성적 처리를 위한 공통 저장소

print('--------------------')
print(' 성적 관리 프로그램 ')
print('--------------------')

# 반복구조를 변형
while True:
    print('--------------------')
    print(' 성적 입력 [ 1 ] ')
    print(' 성적 목록 [ 2 ] ')
    print(' 성적 검색 [ 3 ] ')
    print(' 성적 수정 [ 4 ] ')
    print(' 성적 삭제 [ 5 ] ')
    print(' 성적 종료 [ 6 ] ')
    print('--------------------')


    menu = input('메뉴를 선택 : ')

    if menu == '1':
        student = insert_sungjuk(student)
    elif menu == '2':
        show(student)
    elif menu == '3':
        find_by_name(student)
    elif menu == '4':
        student = update(student)
    elif menu == '5':
        student = delete(student)
    elif menu == '6':
        print('프로그램 종료되었습니다')
        break
    else:
        print('메뉴 선택을 잘못했습니다')
```
